```python

df_subset_without_disease_type = df_subset[cols[1:]]
df_subset_without_disease_type
```

```python
# convert the DataFrame to binary features for all unique values of string data
df_features = pd.get_dummies(df_subset_without_disease_type);
df_features
```

```python
# let's look at the sparsity pattern of the generated features
import matplotlib.pyplot as plt
plt.imshow(df_features, aspect='auto', cmap='binary', interpolation='nearest')
```

```python
# compute a % non-zero for each feature column
import numpy as np 
frac_true = df_features.apply(np.mean);
frac_true
```

```python
# what are the min and max
frac_true.sort_values()
```

```python
# ok, let's keep the ones where at least 1% of entries are different
feature_mask = (frac_true > 0.01) & (frac_true < 0.99);
feature_mask.sum()
```

```python
# which features are we keeping?
feature_mask[feature_mask]
```

```python
df_diverse_features = df_features[df_features.columns[feature_mask]];
df_diverse_features
```

```python
# let's look at the sparsity pattern of the generated features
import matplotlib.pyplot as plt
plt.imshow(df_diverse_features, aspect='auto', cmap='binary', interpolation='nearest')
```

```python
# what are the non-zero fractions in our remaining features
import seaborn as sns
sns.histplot(df_diverse_features.apply(np.mean))
```

```python
import seaborn as sns

for cancer in df_subset.investigation.unique():

    df_for_cancer_type = df_diverse_features.copy()
    label = (df_subset.investigation == cancer)
    corr_vec = df_for_cancer_type.corrwith(label, method='kendall')

    sorted_corr_vec = corr_vec.sort_values()
    print("Top for %9s: (corr=%0.2f) %s" % (cancer, sorted_corr_vec.iloc[-1], sorted_corr_vec.index[-1]))
    f = plt.figure()
    sorted_corr_vec.plot.bar()
    plt.title("Feature correlation with cohort = %s" % cancer)
    f.savefig("Poore-2020-feature-correlation-%s.png" % cancer, bbox_inches='tight')
    plt.close()
    
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score
from sklearn.metrics import RocCurveDisplay


indices = np.arange(len(df_diverse_features))
np.random.shuffle(indices)
train_indices = indices[:len(indices)//2]
test_indices = indices[len(indices)//2:]

X_train = df_diverse_features.values[train_indices]
X_test = df_diverse_features.values[test_indices]

for cancer in df_subset.investigation.unique():
    label = (df_subset.investigation == cancer)
    y_train = label[train_indices]
    y_test = label[test_indices]
    models = {
        "Lasso": LogisticRegression(penalty='l1', solver='liblinear'),
        "Ridge": LogisticRegression(penalty='l2'),
        "RandomForest": RandomForestClassifier(50)
    }
    for model in models.values():
        model.fit(X_train, y_train)
    scores = {k: model.predict_proba(X_test)[:, -1] for (k, model) in models.items()}
    aucs = {k: roc_auc_score(y_test, scores_vec) for (k, scores_vec) in scores.items()}
    print("%10s: %3d pos, L1 AUC=%0.3f, L2 AUC=%0.3f, RF AUC=%0.3f" % (
        cancer, y_train.sum(), aucs["Lasso"], aucs["Ridge"], aucs["RandomForest"]))
    fig, ax = plt.subplots()
    for k, scores_vec in scores.items():
        
        RocCurveDisplay.from_predictions(
            y_test,
            scores_vec,
            name=k,
        
            ax=ax,
        )
    plt.axis("square")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("One-vs-Rest ROC curve: %s" % cancer)
    plt.legend()
    fig.savefig("Poore-2020-predictor-%s-ROC.png" % cancer)
        
```
